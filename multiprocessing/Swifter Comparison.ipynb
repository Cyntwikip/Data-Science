```python
%load_ext autoreload
%autoreload 2

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import functions
import time
import warnings
import seaborn as sns
import multiprocessing as mp

warnings.filterwarnings('ignore')

sns.set_style('darkgrid')

n_cpu = mp.cpu_count()
print("Number of processors: ", n_cpu)
```

```python
import swifter
```

```python
def swifter_approach():
    output = dummy[[0, 1]].swifter.apply(functions.heavy_computation, axis=1)
    
def vanilla_approach():
    with mp.Pool(mp.cpu_count()) as pool:
        output = pool.map(functions.heavy_computation, 
            dummy.itertuples(index=False, name=None)) # name=None fixes pickle issue...
    
def timer(func):
    t_prev = time.time()
    func()
    t_curr = time.time()
    t_dur = t_curr-t_prev 
    t_dur = round(t_dur*1000, 3)
    return t_dur
```

```python
df_vanilla = pd.DataFrame(columns=['size', 'loops', 'time'])
df_swifter = pd.DataFrame(columns=['size', 'loops', 'time'])

for size in [10**i for i in range(1,6)]:
    dummy = pd.DataFrame(np.random.randint(3,10,size=[size, 2]))
    
    for loops in [10**i for i in range(1,4)]:
        functions.loops = loops

        t_dur = timer(vanilla_approach)
        df_vanilla.loc[df_vanilla.shape[0], :] = [size, loops, t_dur]

        t_dur = timer(swifter_approach)
        df_swifter.loc[df_swifter.shape[0], :] = [size, loops, t_dur]
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15,3), dpi=80)

def plot(loops, ax):
    ax = df_vanilla.query(f'loops=={loops}').plot(x='size', y='time', ax=ax, label='Vanilla')
    ax = df_swifter.query(f'loops=={loops}').plot(x='size', y='time', ax=ax, label='Swifter')

    ax.set_title(f'Function Complexity: Loops = {loops}')
    ax.set_xscale('log')
    ax.set_yscale('log')
    ax.set_ylabel('time (ms)')
    ax.legend()
    
plot(10, ax1)
plot(100, ax2)
plot(1000, ax3)
```
